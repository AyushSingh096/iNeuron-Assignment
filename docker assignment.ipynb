## Assignment 2 : Docker

#Solve1 : Demonstrate minimum 15 basic docker command with explanation and screenshot.

1. docker pull krishnaik06/welcomeapp : Using this command we can pull the docker image from dockerhub 
ex:- docker image of welcomeapp from docker hub of krishnaik06 accounts


2. docker images: it will show the docker images available in my account


3.docker run -d -p 5000:5000 krishnaik06/welcomeapp : it will run the docker container krishnaik06/welcomeapp in 
detachble mode with host:5000 and port: 5000 


4.docker ps : it will show all the container running at the moment


5.docker stop 3895aab5a92a : it stop the running container whose id is 3895aab5a92a


6.docker push krishnaik06/welcomeapp:latest : it will push the code to docker hub with version "latest"


7.docker build -t welcome-app . : It will build the docker image welcome-app

8.docker rmi -f welcome-app : it will delete the welcome-app docker image.

9.docker restart welcome-app : It will restart the docker container named as welcome-app

10.docker pause welcome-app: It will pause the welcome-app docker container

11.docker unpause welcome-app: it will unpause or start again the docker container welcome-app

12.docker volume create volume1 : it will create docker volume volume1

13.docker volume rm volume1: It will remove volume1 docker volume

14.docker volume ls: it will list all the docker volume

15.docker network create network1: It will create network1
